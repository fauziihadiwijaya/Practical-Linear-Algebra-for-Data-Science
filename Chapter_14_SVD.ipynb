# Chapter 14 - Singular Value Decomposition (SVD) 🚀
## Compression, Latent Structure, Low-Rank Approximation


## 📘 Overview

Singular Value Decomposition (SVD) is one of the most powerful tools in linear algebra.

Any matrix A can be decomposed as:

A = U Σ Vᵀ

Where:

- U = left singular vectors  
- Σ = singular values  
- Vᵀ = right singular vectors  

SVD is widely used in:

🖼️ Image compression  
🎯 Recommendation systems  
📉 PCA connections  
🔍 Noise reduction  
🤖 Latent semantic analysis


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


## 1️⃣ Compute SVD


```python
A = np.array([
    [3,2,2],
    [2,3,-2]
], dtype=float)

U,S,VT = np.linalg.svd(A)

print("Singular Values:", S)
```

```python
plt.figure(figsize=(6,4))
plt.bar(range(1,len(S)+1), S)
plt.title("Singular Values")
plt.xlabel("Component")
plt.ylabel("Magnitude")
plt.show()
```


## Why Important?

Large singular values capture stronger structure.  
Small singular values often represent noise or weak structure.



## 2️⃣ Matrix Heatmap


```python
plt.figure(figsize=(5,4))
sns.heatmap(A, annot=True, cmap='viridis')
plt.title("Original Matrix")
plt.show()
```


## 3️⃣ Low-Rank Approximation

Keep only top singular components to approximate matrix.


```python
k = 1
Ak = U[:,:k] @ np.diag(S[:k]) @ VT[:k,:]

print(Ak)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.heatmap(A, annot=True, cmap='Blues', cbar=False)
plt.title("Original")

plt.subplot(1,2,2)
sns.heatmap(Ak, annot=True, cmap='Reds', cbar=False)
plt.title("Rank-1 Approximation")

plt.tight_layout()
plt.show()
```


## Why Important?

Low-rank approximation reduces storage and reveals hidden patterns.



## 4️⃣ Image Compression Demo


```python
img = np.random.rand(60,60)

U2,S2,VT2 = np.linalg.svd(img, full_matrices=False)

k = 10
img_k = U2[:,:k] @ np.diag(S2[:k]) @ VT2[:k,:]
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_k, cmap='gray')
plt.title("Compressed Image (k=10)")
plt.axis('off')

plt.tight_layout()
plt.show()
```


## Insight

Many images can be stored approximately using fewer singular components.



## 5️⃣ Reconstruction Error


```python
ks = [1,5,10,20,40]
errors = []

for k in ks:
    rec = U2[:,:k] @ np.diag(S2[:k]) @ VT2[:k,:]
    err = np.linalg.norm(img-rec)
    errors.append(err)

plt.figure(figsize=(7,4))
plt.plot(ks, errors, marker='o')
plt.title("Reconstruction Error vs Rank k")
plt.xlabel("k")
plt.ylabel("Error")
plt.grid(alpha=0.3)
plt.show()
```


As k increases, approximation improves.
