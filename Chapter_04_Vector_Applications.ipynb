## 1️⃣ Cosine Similarity
Measures angle similarity between vectors. Frequently used in NLP, recommendation systems, embeddings.

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3,2])
b = np.array([2,3])

cos_sim = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
print("Cosine Similarity:", round(cos_sim,3))

plt.figure(figsize=(7,7))
plt.quiver(0,0,a[0],a[1],color='blue',angles='xy',scale_units='xy',scale=1,label='a')
plt.quiver(0,0,b[0],b[1],color='green',angles='xy',scale_units='xy',scale=1,label='b')
plt.xlim(0,4); plt.ylim(0,4)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Cosine Similarity Between Two Vectors")
plt.show()
```

## 2️⃣ Correlation
Correlation measures linear relationship between variables.

```python
x = np.arange(10)
y = 2*x + np.random.randn(10)*2

corr = np.corrcoef(x,y)[0,1]
print("Correlation:", round(corr,3))

plt.figure(figsize=(7,5))
plt.scatter(x,y,s=80)
plt.title("Positive Correlation")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha=0.3)
plt.show()
```


## Why Correlation Matters

Used in:

- feature selection  
- statistics  
- exploratory data analysis  
- multicollinearity detection


## 3️⃣ Time Series Filtering
Vectors can represent signals over time. Filters detect trends or smooth noise.

```python
t = np.linspace(0,10,200)
signal = np.sin(2*t) + np.random.randn(200)*0.4

window = np.ones(10)/10
smooth = np.convolve(signal, window, mode='same')

plt.figure(figsize=(10,4))
plt.plot(t,signal,label='Noisy Signal',alpha=0.6)
plt.plot(t,smooth,label='Smoothed Signal',linewidth=3)
plt.title("Moving Average Filter")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```


## Why Important?

Filtering is used in:

- finance trends  
- sensor cleaning  
- biomedical signals  
- preprocessing data


## 4️⃣ Feature Detection with Dot Product
Templates can detect patterns inside signals or images.

```python
template = np.array([1,0,-1])
data = np.array([0,1,2,1,0,-1,-2,-1,0])

scores = np.correlate(data, template, mode='valid')

plt.figure(figsize=(9,4))
plt.stem(scores)
plt.title("Pattern Detection Scores")
plt.grid(alpha=0.3)
plt.show()
```

## 5️⃣ K-Means Clustering
Vectors can represent data points. K-Means groups similar vectors.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X,_ = make_blobs(n_samples=250, centers=3, cluster_std=1.2, random_state=42)

kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=labels,s=40,cmap='viridis')
plt.scatter(centers[:,0],centers[:,1],s=250,c='red',marker='X',label='Centers')
plt.title("K-Means Clustering")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```


## Why K-Means Matters

Used for:

- customer segmentation  
- anomaly grouping  
- image segmentation  
- unsupervised learning


## 6️⃣ Heatmap Similarity Matrix

```python
M = np.random.rand(6,6)
sim = (M + M.T)/2

import seaborn as sns
plt.figure(figsize=(7,5))
sns.heatmap(sim, annot=True, cmap='coolwarm')
plt.title("Similarity Matrix Heatmap")
plt.show()
```
