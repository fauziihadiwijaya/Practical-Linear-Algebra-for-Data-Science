## 1️⃣ What is a General Linear Model?

A linear model assumes:

y = Xβ + ε

Where:

- y = target values  
- X = feature matrix  
- β = coefficients to estimate  
- ε = errors / noise

Goal: find β that best explains the data.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.linspace(0,10,50)
y = 2.5*x + 4 + np.random.randn(50)*3
```


## 2️⃣ Visualizing Data


```python
plt.figure(figsize=(8,5))
plt.scatter(x,y,s=50,alpha=0.7)
plt.title("Observed Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha=0.3)
plt.show()
```


## 3️⃣ Design Matrix

To estimate intercept and slope:

X = [1, x]


```python
X = np.column_stack((np.ones(len(x)), x))
print(X[:5])
```


## 4️⃣ Least Squares Solution

Find coefficients minimizing total squared residual error:

β = argmin ||Xβ - y||²


```python
beta = np.linalg.inv(X.T @ X) @ X.T @ y
print("Intercept =", round(beta[0],3))
print("Slope =", round(beta[1],3))
```


## 📈 Best Fit Line


```python
y_pred = X @ beta

plt.figure(figsize=(8,5))
plt.scatter(x,y,alpha=0.6,label='Data')
plt.plot(x,y_pred,color='red',linewidth=3,label='Best Fit Line')
plt.title("Least Squares Regression")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```


## Why Squared Errors?

Squaring errors:

✅ penalizes large mistakes  
✅ differentiable for optimization  
✅ gives elegant matrix solution



## 5️⃣ Residual Errors

Residual = actual - predicted


```python
res = y - y_pred

plt.figure(figsize=(8,5))
plt.scatter(y_pred,res,alpha=0.7)
plt.axhline(0,color='red',linestyle='--')
plt.title("Residual Plot")
plt.xlabel("Predicted")
plt.ylabel("Residual")
plt.grid(alpha=0.3)
plt.show()
```


## Why Residual Plot Matters?

Used to detect:

- nonlinearity  
- unequal variance  
- outliers  
- bad model fit



## 6️⃣ Geometric Interpretation

Least squares projects y onto the column space of X.

This is a geometric projection problem, not only statistics.


```python
import seaborn as sns

XtX = X.T @ X

plt.figure(figsize=(6,5))
sns.heatmap(XtX, annot=True, cmap='viridis')
plt.title("XᵀX Matrix")
plt.show()
```


## 7️⃣ Why Important in Data Science?
