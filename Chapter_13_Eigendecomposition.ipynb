# Chapter 13 - Eigendecomposition 🚀
## Eigenvalues, Eigenvectors, Dominant Directions


## 📘 Overview

Eigendecomposition is one of the most important concepts in linear algebra.

It reveals special directions that remain unchanged in direction after a transformation.

For matrix A:

A v = λ v

Where:

- v = eigenvector  
- λ = eigenvalue

This chapter is fundamental for:

📉 PCA  
📊 Covariance analysis  
🤖 Machine learning  
🔁 Dynamical systems  
🧠 Deep learning optimization



## 1️⃣ Intuition

Most vectors change direction after matrix multiplication.

But some rare vectors only get stretched or compressed.

Those are eigenvectors.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

A = np.array([[3,1],
              [0,2]])

vals, vecs = np.linalg.eig(A)

print("Eigenvalues:\n", vals)
print("\nEigenvectors:\n", vecs)
```


## 2️⃣ Matrix Heatmap


```python
plt.figure(figsize=(5,4))
sns.heatmap(A, annot=True, cmap='viridis')
plt.title("Matrix A")
plt.show()
```


## 3️⃣ Eigenvector Directions


```python
plt.figure(figsize=(7,7))

colors = ['blue', 'red', 'green', 'purple', 'orange']

for i in range(vecs.shape[1]):
    v = vecs[:,i].real

    plt.quiver(
        0, 0,
        v[0], v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        color=colors[i],
        width=0.008,
        label=f'Eigenvector {i+1}'
    )

    plt.text(
        v[0]*1.08,
        v[1]*1.08,
        f'v{i+1}',
        color=colors[i],
        fontsize=12,
        weight='bold'
    )

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.axhline(0,color='gray',linewidth=0.8)
plt.axvline(0,color='gray',linewidth=0.8)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Eigenvector Directions", fontsize=14, weight='bold')
plt.show()
```


## Why Important?

Eigenvectors reveal stable / principal directions in data.



## 4️⃣ Transformation of Eigenvectors


```python
plt.figure(figsize=(8,7))

colors = ['blue','red']

for i in range(vecs.shape[1]):
    v = vecs[:,i].real
    Av = A @ v

    plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,
               color=colors[i],label=f'v{i+1}')
    plt.quiver(0,0,Av[0],Av[1],angles='xy',scale_units='xy',scale=1,
               color=colors[i],alpha=0.4,label=f'Av{i+1}')

plt.xlim(-2,3)
plt.ylim(-2,3)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Eigenvectors Keep Direction")
plt.show()
```


Notice that transformed vectors stay on same lines, only scaled.



## 5️⃣ Covariance Matrix and PCA Connection


```python
np.random.seed(42)

X = np.random.multivariate_normal(
    mean=[0,0],
    cov=[[4,3],[3,3]],
    size=300
)

C = np.cov(X.T)
evals,evecs = np.linalg.eig(C)
```

```python
plt.figure(figsize=(7,6))
plt.scatter(X[:,0],X[:,1],alpha=0.4)

for i in range(2):
    v = evecs[:,i]
    plt.quiver(0,0,v[0]*3,v[1]*3,angles='xy',scale_units='xy',scale=1,
               linewidth=3,label=f'PC Direction {i+1}')

plt.grid(alpha=0.3)
plt.legend()
plt.title("Covariance Eigenvectors")
plt.show()
```


## Why PCA Uses Eigenvectors

Principal components are eigenvectors of covariance matrix.

Largest eigenvalue = direction of maximum variance.



## 6️⃣ Eigenvalues as Importance Scores


```python
plt.figure(figsize=(7,4))
plt.bar(['λ1','λ2'], evals)
plt.title("Eigenvalues from Covariance Matrix")
plt.ylabel("Magnitude")
plt.show()
```
