# 1️⃣ Linear Combination

A **linear combination** means combining vectors using scalar weights.

Mathematically:

v = a₁x₁ + a₂x₂ + ... + aₙxₙ

This means we scale vectors, then add them together.

## Why Important?

Almost everything in linear algebra is built from linear combinations:

- predictions in regression  
- neural network weighted sums  
- PCA components  
- coordinate systems  
- basis representation  

If you understand linear combinations, many future topics become easier.


```python
import numpy as np

x1 = np.array([1,0])
x2 = np.array([0,1])

v = 3*x1 + 2*x2
print("Linear Combination Result:", v)
```


# 2️⃣ Visualization of Linear Combination
Using basis vectors to build new vectors.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,7))

plt.quiver(0,0,1,0,color='blue',angles='xy',scale_units='xy',scale=1,label='x1')
plt.quiver(0,0,0,1,color='green',angles='xy',scale_units='xy',scale=1,label='x2')
plt.quiver(0,0,3,2,color='red',angles='xy',scale_units='xy',scale=1,label='3x1 + 2x2')

plt.xlim(0,4)
plt.ylim(0,4)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Linear Combination")
plt.show()
```


# 3️⃣ Span

The **span** of vectors is the set of all possible linear combinations formed from those vectors.

Examples:

- Two independent vectors in 2D usually span the entire plane.
- One vector spans only a line.
- Redundant vectors may not increase span.

## Why Important?

Span tells us what outputs are possible.

In machine learning, span relates to:

- representational power  
- feature coverage  
- subspaces in PCA  
- solution spaces



# 4️⃣ Linear Independence

A set of vectors is **linearly independent** if no vector can be created from the others.

If one vector can be written as a combination of others, then the set is dependent.

## Why Important?

Independent vectors contain unique information.

Dependent vectors often create problems:

❌ redundant features  
❌ unstable regression coefficients  
❌ singular matrices  
❌ wasted dimensions

This concept is extremely important in statistics and ML.


```python
a = np.array([1,2])
b = np.array([2,4])

print("b is multiple of a -> dependent")
```


# 5️⃣ Basis

A **basis** is a set of vectors that is:

✅ linearly independent  
✅ spans the space

This means a basis is the minimum efficient set of vectors needed to describe a space.

Examples:

- Standard basis in 2D:
e₁ = [1,0]
e₂ = [0,1]

## Why Important?

Basis is everywhere:

- coordinate systems  
- PCA creates new basis vectors  
- Fourier transforms use basis functions  
- latent feature models use learned bases


```python
e1 = np.array([1,0])
e2 = np.array([0,1])

point = 4*e1 + 3*e2
print("Point from basis:", point)
```


# 6️⃣ Common Beginner Mistakes

❌ Thinking more vectors always means more information  
❌ Ignoring redundancy  
❌ Confusing span with magnitude  
❌ Memorizing basis definition without intuition  
❌ Not connecting independence to real datasets



# 7️⃣ Why This Matters for Data Science

Suppose dataset has features:

- salary  
- monthly income  

These may be strongly dependent.

That means duplicated information.

Models may suffer from multicollinearity.

Good feature engineering often means finding a better basis with less redundancy.
