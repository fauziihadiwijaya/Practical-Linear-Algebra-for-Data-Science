## 1️⃣ Matrix Rank

The **rank** of a matrix tells us how many independent rows or columns it has.

It measures how much unique information exists inside the matrix.

### Why Important?

- Detect redundant features  
- Determine invertibility  
- Understand dimensionality  
- Important in PCA and regression


```python
import numpy as np

A = np.array([[1,2],
              [2,4]])

B = np.array([[1,2],
              [3,4]])

print("Rank of A:", np.linalg.matrix_rank(A))
print("Rank of B:", np.linalg.matrix_rank(B))
```

## 📊 Rank Heatmap Comparison

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.heatmap(A, annot=True, cbar=False)
plt.title("Low Rank Matrix")

plt.subplot(1,2,2)
sns.heatmap(B, annot=True, cbar=False)
plt.title("Full Rank Matrix")

plt.tight_layout()
plt.show()
```


## 2️⃣ Determinant

The determinant measures area/volume scaling caused by a matrix transformation.

If determinant = 0:

❌ space collapses  
❌ inverse does not exist  
❌ information lost


```python
M = np.array([[2,1],
              [1,2]])

det = np.linalg.det(M)
print("Determinant:", round(det,3))
```

## 🔺 Geometric Determinant Visualization

```python
square = np.array([[0,0],[1,0],[1,1],[0,1],[0,0]])
trans = square @ M.T

plt.figure(figsize=(7,7))
plt.plot(square[:,0],square[:,1],label='Original Square')
plt.plot(trans[:,0],trans[:,1],label='Transformed Shape')
plt.grid(alpha=0.3)
plt.legend()
plt.title("Determinant = Area Scaling")
plt.show()
```


## 3️⃣ Column Space

The column space is the set of all linear combinations of matrix columns.

It tells us what outputs can be produced by Ax.


```python
c1 = np.array([1,1])
c2 = np.array([2,1])

plt.figure(figsize=(7,7))
plt.quiver(0,0,c1[0],c1[1],color='blue',angles='xy',scale_units='xy',scale=1,label='col1')
plt.quiver(0,0,c2[0],c2[1],color='green',angles='xy',scale_units='xy',scale=1,label='col2')
plt.xlim(0,3); plt.ylim(0,3)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Column Space Generators")
plt.show()
```


## Why Important?

Used in solving systems:

Ax = b

If b is outside column space, no exact solution exists.



## 4️⃣ Null Space

The null space contains vectors x such that:

Ax = 0

These inputs disappear after transformation.


```python
N = np.array([[1,1],
              [2,2]])

u = np.array([1,-1])

print("A @ u =", N @ u)
```

## 🕳️ Null Space Visual

```python
plt.figure(figsize=(7,6))
x = np.linspace(-3,3,50)
y = -x

plt.plot(x,y,label='Null Space Line')
plt.grid(alpha=0.3)
plt.legend()
plt.title("Vectors Producing Zero Output")
plt.show()
```


## 5️⃣ Matrix Norm

A matrix norm measures overall matrix size / magnitude.

Useful in:

- optimization  
- regularization  
- numerical stability  
- comparing matrices


```python
X = np.random.randn(4,4)
fro = np.linalg.norm(X)

print("Frobenius Norm:", round(fro,3))
```

## 🌡️ Random Matrix Heatmap

```python
plt.figure(figsize=(6,5))
sns.heatmap(X, cmap='coolwarm', annot=True)
plt.title("Matrix Used for Norm")
plt.show()
```


## 6️⃣ Common Beginner Mistakes
