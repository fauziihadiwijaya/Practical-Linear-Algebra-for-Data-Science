## 1️⃣ Covariance Matrix

A covariance matrix shows how variables change together.

Positive covariance:
- both increase together

Negative covariance:
- one increases while the other decreases

Near zero:
- weak linear relationship


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

X = np.random.multivariate_normal(
    mean=[0,0],
    cov=[[3,2],[2,2]],
    size=300
)

cov_mat = np.cov(X.T)
print(cov_mat)
```

```python
plt.figure(figsize=(7,6))
plt.scatter(X[:,0],X[:,1],alpha=0.6)
plt.title("Two Correlated Variables")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(cov_mat, annot=True, cmap='coolwarm')
plt.title("Covariance Matrix")
plt.show()
```


## Why Important?

Covariance matrices are essential in:

- PCA  
- portfolio risk analysis  
- multivariate statistics  
- anomaly detection



## 2️⃣ Correlation Matrix

Correlation normalizes covariance into range [-1,1].


```python
corr = np.corrcoef(X.T)

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='viridis', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```


## 3️⃣ Geometric Transformation with Matrix

Matrices can rotate, scale, reflect, and shear points in space.


```python
pts = np.array([
    [1,1],
    [2,1],
    [2,2],
    [1,2],
    [1,1]
])

theta = np.pi/4
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

rot = pts @ R.T
```

```python
plt.figure(figsize=(7,7))
plt.plot(pts[:,0],pts[:,1],label='Original Shape')
plt.plot(rot[:,0],rot[:,1],label='Rotated Shape')
plt.grid(alpha=0.3)
plt.legend()
plt.title("Matrix Rotation Transformation")
plt.show()
```


## Why Important?

Used in:

- computer graphics  
- robotics  
- coordinate systems  
- PCA rotations



## 4️⃣ Image Feature Detection

Images are matrices. Small kernels can detect edges and patterns.


```python
img = np.zeros((20,20))
img[:,10:] = 1

kernel = np.array([
    [-1,1]
])

feature = np.zeros((20,19))
for i in range(20):
    for j in range(19):
        feature[i,j] = np.sum(img[i,j:j+2]*kernel)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(feature, cmap='coolwarm')
plt.title("Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()
```


## Why Important?

This is the foundation of:

- image processing  
- convolutional neural networks  
- object detection



## 5️⃣ Pairwise Distance Matrix

Many ML algorithms compare distances between samples.


```python
from sklearn.metrics import pairwise_distances

sample = X[:10]
D = pairwise_distances(sample)

plt.figure(figsize=(7,5))
sns.heatmap(D, cmap='magma')
plt.title("Pairwise Distance Matrix")
plt.show()
```
