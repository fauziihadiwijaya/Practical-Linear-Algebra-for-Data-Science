# Chapter 02 - Vectors Part 1  🚀

## 📘 Overview
Vectors are core objects in linear algebra used for representing data, direction, features, and relationships.

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3,4])
w = np.array([1,2])
u = v + w
```

## 🎯 Colored 2D Vector Visualization

```python
plt.figure(figsize=(7,7))

plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='blue',label='v')
plt.quiver(0,0,w[0],w[1],angles='xy',scale_units='xy',scale=1,color='green',label='w')
plt.quiver(0,0,u[0],u[1],angles='xy',scale_units='xy',scale=1,color='red',label='v+w')

plt.text(v[0]+0.1,v[1],'v',fontsize=12,color='blue')
plt.text(w[0]+0.1,w[1],'w',fontsize=12,color='green')
plt.text(u[0]+0.1,u[1],'v+w',fontsize=12,color='red')

plt.xlim(0,6)
plt.ylim(0,7)
plt.grid(alpha=0.3)
plt.legend()
plt.title("2D Vectors with Different Colors")
plt.show()
```

## 📏 Magnitude of Vector

```python
print("||v|| =", np.linalg.norm(v))
print("||w|| =", np.linalg.norm(w))
```

## 🔥 Dot Product

```python
dot = np.dot(v,w)
print("v · w =", dot)
```

## 📐 Projection of v onto w

```python
proj = (np.dot(v,w)/np.dot(w,w))*w
print("Projection:", proj)

plt.figure(figsize=(7,7))
plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='blue',label='v')
plt.quiver(0,0,w[0],w[1],angles='xy',scale_units='xy',scale=1,color='green',label='w')
plt.quiver(0,0,proj[0],proj[1],angles='xy',scale_units='xy',scale=1,color='purple',label='proj(v onto w)')
plt.plot([v[0],proj[0]],[v[1],proj[1]],'--',alpha=0.6)

plt.xlim(0,6)
plt.ylim(0,7)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Vector Projection")
plt.show()
```

## 🌌 3D Vector Visualization

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

a = np.array([2,1,3])
b = np.array([1,3,2])

ax.quiver(0,0,0,a[0],a[1],a[2],color='blue')
ax.quiver(0,0,0,b[0],b[1],b[2],color='orange')

ax.text(a[0],a[1],a[2],'a',color='blue')
ax.text(b[0],b[1],b[2],'b',color='orange')

ax.set_xlim([0,4]); ax.set_ylim([0,4]); ax.set_zlim([0,4])
ax.set_title("3D Vectors")
plt.show()
```

## 🌟 Summary
Different colors improve readability and intuition. Vectors help represent data, similarity, and direction in machine learning.
