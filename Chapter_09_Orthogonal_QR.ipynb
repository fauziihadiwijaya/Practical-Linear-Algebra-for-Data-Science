## 📘 Overview

Orthogonality is one of the most beautiful and useful ideas in linear algebra.

Two vectors are orthogonal when they meet at 90° and their dot product is zero.

Orthogonal matrices preserve:

📏 lengths  
📐 angles  
🔄 geometry  

This chapter also introduces **QR Decomposition**, a powerful factorization used in least squares and numerical computing.



## 1️⃣ Orthogonal Vectors

If:

u · v = 0

then vectors are orthogonal.


```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([3,0])
v = np.array([0,2])

print("Dot product =", np.dot(u,v))
```

```python
plt.figure(figsize=(7,7))
plt.quiver(0,0,u[0],u[1],color='blue',angles='xy',scale_units='xy',scale=1,label='u')
plt.quiver(0,0,v[0],v[1],color='green',angles='xy',scale_units='xy',scale=1,label='v')
plt.xlim(0,4); plt.ylim(0,3)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Orthogonal Vectors")
plt.show()
```


## Why Important?

Used in:

- projections  
- PCA directions  
- signal separation  
- numerical stability



## 2️⃣ Orthogonal Matrix

A matrix Q is orthogonal if:

QᵀQ = I

Columns of Q are orthonormal vectors.


```python
theta = np.pi/4

Q = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

print(Q)
print("\nQ.T @ Q =\n", Q.T @ Q)
```

## 🔄 Rotation Matrix Visualization

```python
pts = np.array([
    [1,0],[2,0],[2,1],[1,1],[1,0]
])

rot = pts @ Q.T

plt.figure(figsize=(7,7))
plt.plot(pts[:,0],pts[:,1],label='Original')
plt.plot(rot[:,0],rot[:,1],label='Rotated')
plt.grid(alpha=0.3)
plt.legend()
plt.title("Orthogonal Matrix Preserves Shape")
plt.show()
```


## 3️⃣ Gram-Schmidt Process

Converts independent vectors into orthogonal basis vectors.

Very important because orthogonal vectors are easier to work with.


```python
a = np.array([1.,1.])
b = np.array([1.,0.])

e1 = a / np.linalg.norm(a)

proj = np.dot(b,e1)*e1
u2 = b - proj
e2 = u2 / np.linalg.norm(u2)

print("e1 =", e1)
print("e2 =", e2)
print("Dot =", np.dot(e1,e2))
```

```python
plt.figure(figsize=(7,7))
plt.quiver(0,0,e1[0],e1[1],color='blue',angles='xy',scale_units='xy',scale=1,label='e1')
plt.quiver(0,0,e2[0],e2[1],color='red',angles='xy',scale_units='xy',scale=1,label='e2')
plt.xlim(-1,1.5); plt.ylim(-1,1.5)
plt.grid(alpha=0.3)
plt.legend()
plt.title("Gram-Schmidt Orthogonal Basis")
plt.show()
```


## 4️⃣ QR Decomposition

Any full-rank matrix A can be decomposed:

A = QR

Where:

Q = orthogonal matrix  
R = upper triangular matrix


```python
A = np.array([[1,2],
              [3,4],
              [5,6]])

Q,R = np.linalg.qr(A)

print("Q shape:", Q.shape)
print("R shape:", R.shape)
```

```python
import seaborn as sns

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.heatmap(Q, annot=True, cmap='coolwarm')
plt.title("Q Matrix")

plt.subplot(1,2,2)
sns.heatmap(R, annot=True, cmap='viridis')
plt.title("R Matrix")

plt.tight_layout()
plt.show()
```


## Why QR Matters?

Used in:

- least squares regression  
- solving linear systems  
- stable computations  
- eigenvalue algorithms



## 5️⃣ Geometric Intuition

Q rotates/reflection transforms space.  
R scales/shears coordinates.
