# Chapter 08 - Matrix Inverse 🚀
## Invertibility, Solving Systems, Pseudoinverse


## 📘 Overview

The matrix inverse is one of the most famous ideas in linear algebra.

For numbers:

a × a⁻¹ = 1

For matrices:

A A⁻¹ = I

where **I** is the identity matrix.

Inverse matrices are used to reverse transformations and solve systems of equations.

This chapter also explains why inverse is not always possible or numerically stable.



## 1️⃣ Identity Matrix

The identity matrix acts like number 1 in matrix multiplication.


```python
import numpy as np

I = np.eye(3)
print(I)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,4))
sns.heatmap(I, annot=True, cmap='Blues', cbar=False)
plt.title("Identity Matrix")
plt.show()
```


## 2️⃣ Computing the Inverse

If matrix A is invertible:

A⁻¹ exists


```python
A = np.array([[2,1],
              [1,1]])

A_inv = np.linalg.inv(A)

print("A =\n", A)
print("\nA inverse =\n", A_inv)
```

```python
print("A @ A_inv =\n", A @ A_inv)
```

## 🔥 Heatmap of Inverse

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.heatmap(A, annot=True, cbar=False, cmap='viridis')
plt.title("Matrix A")

plt.subplot(1,2,2)
sns.heatmap(A_inv, annot=True, cbar=False, cmap='coolwarm')
plt.title("Inverse A⁻¹")

plt.tight_layout()
plt.show()
```


## 3️⃣ Geometric Interpretation

If A stretches / rotates space, then A⁻¹ undoes that transformation.


```python
pts = np.array([
    [0,0],[1,0],[1,1],[0,1],[0,0]
])

trans = pts @ A.T
recover = trans @ A_inv.T

plt.figure(figsize=(7,7))
plt.plot(pts[:,0],pts[:,1],label='Original')
plt.plot(trans[:,0],trans[:,1],label='Transformed')
plt.plot(recover[:,0],recover[:,1],'--',label='Recovered')
plt.grid(alpha=0.3)
plt.legend()
plt.title("Transformation and Inverse Recovery")
plt.show()
```


## Why Important?

Used in:

- solving linear systems  
- coordinate reversal  
- graphics  
- regression derivations  
- optimization theory



## 4️⃣ Singular Matrix (No Inverse)

If determinant = 0, inverse does not exist.


```python
S = np.array([[1,2],
              [2,4]])

print("Determinant:", np.linalg.det(S))
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(S, annot=True, cmap='Reds', cbar=False)
plt.title("Singular Matrix")
plt.show()
```


Rows are dependent, so information collapses.



## 5️⃣ Solving Systems Ax = b

Instead of computing inverse manually:

x = A⁻¹b


```python
b = np.array([5,3])

x = np.linalg.solve(A,b)

print("Solution x =", x)
```


## Better Practice

Use `np.linalg.solve()` instead of explicit inverse when possible.
It is faster and more stable.



## 6️⃣ Moore-Penrose Pseudoinverse

Used when matrix is not square or not invertible.
Very important in machine learning.


```python
R = np.array([[1,2],
              [3,4],
              [5,6]])

pinv = np.linalg.pinv(R)

print("Pseudoinverse shape:", pinv.shape)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(pinv, annot=True, cmap='magma')
plt.title("Pseudoinverse")
plt.show()
```


## Why Important?

Used in:

- least squares regression  
- overdetermined systems  
- linear models  
- recommendation systems



## 7️⃣ Common Beginner Mistakes

❌ Thinking every matrix has inverse  
❌ Using inverse when solve() is better  
❌ Ignoring numerical instability  
❌ Forgetting determinant condition
