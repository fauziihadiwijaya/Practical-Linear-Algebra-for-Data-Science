```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris, make_blobs
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```


## 1️⃣ PCA Intuition

PCA finds new orthogonal directions capturing maximum variance.

Benefits:

- reduce dimensions  
- remove redundancy  
- faster training  
- visualization


```python
iris = load_iris()
X = iris.data
y = iris.target

pca = PCA(n_components=2)
Xp = pca.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(Xp[:,0], Xp[:,1], c=y, cmap='viridis', s=50)
plt.title("PCA Projection of Iris Dataset")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.bar(['PC1','PC2'], pca.explained_variance_ratio_)
plt.title("Explained Variance Ratio")
plt.ylabel("Variance Share")
plt.show()
```


## Why Important?

PCA keeps most information using fewer variables.



## 2️⃣ LDA Intuition

LDA seeks directions maximizing class separation.
Unlike PCA, it uses labels.


```python
lda = LinearDiscriminantAnalysis(n_components=2)
Xl = lda.fit_transform(X,y)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(Xl[:,0], Xl[:,1], c=y, cmap='plasma', s=50)
plt.title("LDA Projection of Iris Dataset")
plt.xlabel("LD1")
plt.ylabel("LD2")
plt.grid(alpha=0.3)
plt.show()
```


## PCA vs LDA

PCA = maximize variance  
LDA = maximize class separation



## 3️⃣ PCA vs LDA Side by Side


```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(Xp[:,0],Xp[:,1],c=y,cmap='viridis')
plt.title("PCA")

plt.subplot(1,2,2)
plt.scatter(Xl[:,0],Xl[:,1],c=y,cmap='plasma')
plt.title("LDA")

plt.tight_layout()
plt.show()
```


## 4️⃣ SVD for Recommendation Systems

User-item ratings can be represented as a matrix.
SVD finds hidden preference factors.


```python
R = np.array([
    [5,4,0,1],
    [4,0,0,1],
    [1,1,0,5],
    [1,0,0,4],
    [0,1,5,4]
], dtype=float)

U,S,VT = np.linalg.svd(R, full_matrices=False)
```

```python
plt.figure(figsize=(6,4))
plt.bar(range(1,len(S)+1), S)
plt.title("Singular Values of Ratings Matrix")
plt.xlabel("Latent Factor")
plt.ylabel("Strength")
plt.show()
```

```python
R2 = U[:,:2] @ np.diag(S[:2]) @ VT[:2,:]

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.heatmap(R, annot=True, cmap='Blues')
plt.title("Original Ratings")

plt.subplot(1,2,2)
sns.heatmap(R2, annot=True, cmap='Oranges')
plt.title("Low-Rank Recommendation Approximation")

plt.tight_layout()
plt.show()
```


## Why Important?

Used in recommendation engines like:

- movies  
- music  
- e-commerce products



## 5️⃣ Common Beginner Mistakes

❌ Using PCA without scaling features  
❌ Expecting PCA to optimize class separation  
❌ Using LDA without labels  
❌ Ignoring interpretation of latent factors
