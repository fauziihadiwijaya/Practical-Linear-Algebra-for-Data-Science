## 1️⃣ Systems of Equations as Matrices

Example:

x + y = 5  
2x + y = 8

Can be written:

Ax = b


```python
import numpy as np

A = np.array([[1,1],
              [2,1]], dtype=float)

b = np.array([5,8], dtype=float)

x = np.linalg.solve(A,b)

print("Solution [x,y] =", x)
```


## Why Important?

Matrix form scales to thousands or millions of equations.



## 2️⃣ Augmented Matrix Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

Aug = np.column_stack((A,b))

plt.figure(figsize=(6,4))
sns.heatmap(Aug, annot=True, cmap='Blues', cbar=False)
plt.title("Augmented Matrix [A|b]")
plt.show()
```


## 3️⃣ Row Reduction

We use row operations to simplify the system.

Allowed operations:

- swap rows  
- multiply row by scalar  
- add multiple of one row to another


```python
R = Aug.copy()

# eliminate below pivot
R[1] = R[1] - 2*R[0]
print(R)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.heatmap(Aug, annot=True, cbar=False)
plt.title("Before Elimination")

plt.subplot(1,2,2)
sns.heatmap(R, annot=True, cbar=False)
plt.title("After Elimination")

plt.tight_layout()
plt.show()
```


## 4️⃣ Gaussian Elimination

Transforms matrix into upper triangular form, then uses back substitution.


```python
U = np.array([[1,1,5],
              [0,-1,-2]], dtype=float)

y = U[1,2]/U[1,1]
x = (U[0,2] - U[0,1]*y)/U[0,0]

print("x =", x)
print("y =", y)
```


## Why Important?

Foundation of many linear solvers used in software.



## 5️⃣ LU Decomposition

Factor matrix into:

A = LU

Where:

L = lower triangular  
U = upper triangular


```python
from scipy.linalg import lu

P,L,U = lu(A)

print("L =\n", L)
print("\nU =\n", U)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.heatmap(L, annot=True, cmap='Greens', cbar=False)
plt.title("Lower Triangular L")

plt.subplot(1,2,2)
sns.heatmap(U, annot=True, cmap='Oranges', cbar=False)
plt.title("Upper Triangular U")

plt.tight_layout()
plt.show()
```


## Why LU Matters?

Instead of solving Ax=b repeatedly from scratch:

1. Factor A once into LU  
2. Solve quickly for many different b vectors

Used heavily in engineering and simulation.



## 6️⃣ Permutation Matrix

Sometimes rows must be swapped for numerical stability.

This creates:

PA = LU


```python
print("Permutation Matrix P =\n", P)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(P, annot=True, cmap='Purples', cbar=False)
plt.title("Permutation Matrix")
plt.show()
```


## 7️⃣ Common Beginner Mistakes

❌ Thinking equations and matrices are separate topics  
❌ Forgetting row operations preserve solutions  
❌ Ignoring pivot problems  
❌ Memorizing LU without purpose
